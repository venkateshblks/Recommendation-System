# Feature Extraction:

```python
anime1['features'] = anime1['genre'] + ' ' + anime1['rating'].astype(str)
anime1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
genre_matrix = vectorizer.fit_transform(anime1['features'])
genre_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(genre_matrix)
k=pd.DataFrame(similarity_matrix)
k
```

```python
k.index=anime1.anime_id
k.columns=anime1.anime_id
k
```

```python
np.fill_diagonal(similarity_matrix,0)
k
```

# Recommendation System:

```python
k.idxmax(axis=1)[0:5]
```

```python
anime1[(anime1['anime_id']==9969) | (anime1['anime_id']==15335)]
```

```python
def recommend_anime(anime_id, top_n=10,similarity_threshold=0.2):
    anime_idx = anime1[anime1['anime_id'] == anime_id].index[0]

    sim_scores = list(enumerate(similarity_matrix[anime_idx]))
    filter_sim_scores = [x for x in sim_scores if x[1] >= similarity_threshold]
    filter_sim_scores = sorted(filter_sim_scores, key=lambda x: x[1], reverse=True)
    print('length of all similarity scores :',len(filter_sim_scores))
    filter_sim_scores = filter_sim_scores[1:top_n + 1]
    recommended_anime_ids = [anime1.iloc[i[0]]['anime_id'] for i in filter_sim_scores]

    print(f'similarity scores of top {top_n} :',filter_sim_scores )
    return recommended_anime_ids
```

```python
recommended_anime = recommend_anime(anime_id=9969, top_n=5)
print(recommended_anime)
anime1[anime1['anime_id'].isin(recommended_anime)]
```

```python
recommended_anime = recommend_anime(anime_id=9969, top_n=5,similarity_threshold=0.5)
print(recommended_anime)
anime1[anime1['anime_id'].isin(recommended_anime)]
```

```python
recommended_anime = recommend_anime(anime_id=15335, top_n=10,similarity_threshold=0.5)
print(recommended_anime)
anime1[anime1['anime_id'].isin(recommended_anime)]
```

1. **Can you explain the difference between user-based and item-based collaborative filtering?**


In user-based collaborative filtering, the system recommends items to a user based on the preferences and behaviors of other users who are similar to them

In item-based collaborative filtering, the system recommends items to a user based on the similarities between items.

**2. What is collaborative filtering, and how does it work?**

Collaborative Filtering is a popular approach in recommendation systems that predicts a user's preferences or interests based on the preferences and behavior of similar users or based on the similarity between items. The method leverages patterns in user interactions, such as ratings, purchases, or other forms of feedback, to make personalized recommendations.

The method calculates the similarity between users or between items. This can be done using metrics such as cosine similarity, Pearson correlation, or other measures.

Once similarities are computed, the system uses this information to recommend items to a target user. This can be done in two ways:

User-based: Recommend items to a user based on the preferences of similar users.  
Item-based: Recommend items to a user based on their past interactions and the similarity between items.  
The system ranks and filters recommendations based on similarity scores, user preferences, or other criteria such as diversity and novelty
